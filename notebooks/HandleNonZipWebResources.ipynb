```python
import os
from bs4 import BeautifulSoup
from ricecooker.utils.downloader import download_static_assets, make_request
from ricecooker.utils.zip import create_predictable_zip
from urllib.parse import urljoin
import tempfile

url = r1['url']
print(url)
# base_url = 'http://www.prathamopenschool.org/CourseContent/Games/Mathematics/'



    
def download_html_subpage_and_resources(referring_url, rel_path, dest_dir):
    """
    Download additional HTML file at `rel_path` and its associated assets
    to `dest_dir`. We assume link `rel_path` occurs on page `referring_url`.
    """
    print("   in download_html_subpage_and_resources for", rel_path)
    url = urljoin(referring_url,rel_path)
    print('   resolved to url')
    response = make_request(url)
    response.encoding = 'utf-8'
    html = response.text
    doc = BeautifulSoup(html, "html.parser")
    html_pr = download_static_assets(doc, dest_dir, url)
    with open( os.path.join(dest_dir, rel_path), 'w') as htmlf:
        htmlf.write(str(html_pr))

    # recurse on links within subpage
    links = doc.find('body').find_all('a')
    for link in links:
        rel_path = link['href']
        # avoid re-downloading things we've already gotten
        if not os.path.exists(os.path.join(dest_dir, rel_path)):
            download_html_subpage_and_resources(url, rel_path, dest_dir)
    

    
def downlaod_UnsupportedPrathamWebResource(url):
    print("Downloading UnsupportedPrathamWebResource", url)

    # store here:
    dest_dir = tempfile.mkdtemp()
    print('storing files in', dest_dir)
    
    response = make_request(url)
    response.encoding = 'utf-8'
    html = response.text
    doc = BeautifulSoup(html, "html.parser")
    html_pr = download_static_assets(doc, dest_dir, url)
    with open(os.path.join(dest_dir, 'index.html'), 'w') as indexf:
        indexf.write(str(html_pr))
    links = doc.find('body').find_all('a')
    for link in links:
        rel_path = link['href']
        download_html_subpage_and_resources(url, rel_path, dest_dir)

    zip_path = create_predictable_zip(dest_dir)
    return zip_path


downlaod_UnsupportedPrathamWebResource(url)
```
