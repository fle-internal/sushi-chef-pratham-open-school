# Multi-langage game name logic

We use `GameName` to refer to a game, but the actual faile name is `GameName_HI`, `GameName_MR`, etc.
Let's see how this works.


```python
import json
from sushichef import PRADIGI_WEBSITE_LANGUAGES
```

## Game extraction

```python
from sushichef import extract_website_games_from_tree

# This code runs after all languages have been crawled
website_games = {}
for lang in PRADIGI_WEBSITE_LANGUAGES:
    lang_games = extract_website_games_from_tree(lang)
    website_games[lang] = lang_games
WEBSITE_GAMES_OUTPUT = 'chefdata/trees/website_games_all_langs.json'
# Save website games
with open(WEBSITE_GAMES_OUTPUT, 'w') as json_file:
    json.dump(website_games, json_file, ensure_ascii=False, indent=2, sort_keys=True)
```
